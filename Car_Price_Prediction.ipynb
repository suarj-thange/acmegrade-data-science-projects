```python
X = df.iloc[:, [0,1,3,4,5,6,7,8]].values
Y = df.iloc[:, [2]].values

X.shape, Y.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le1 = LabelEncoder()
X[:, 0] = le1.fit_transform(X[:, 0])   # model column

le2 = LabelEncoder()
X[:, -4] = le2.fit_transform(X[:, -4])  # fuelType column

X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [2])],
    remainder='passthrough'
)

X = ct.fit_transform(X)
X.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)

X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=0
)

X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, Y_train)
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

print("R2 Score:", r2_score(Y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(Y_test, y_pred))
```

```python
import pandas as pd

comparison = pd.concat([
    pd.DataFrame(Y_test, columns=['Actual Price']),
    pd.DataFrame(y_pred, columns=['Predicted Price'])
], axis=1)

comparison.head(10)
```

```python
full_pred = rf.predict(X)
full_pred
```

```python
print("R2 Score (Full Data):", r2_score(Y, full_pred))
print("MAE (Full Data):", mean_absolute_error(Y, full_pred))
```
